## Adjusting the Plot: Line Colors and Styles

```python
# using seaborn palette
with sns.color_palette("viridis"):
    for phase in range(6):
        plt.plot(x,np.sin(x-phase))
```

```python
# another ways of styling
plt.plot(x,np.sin(x-0),color = 'blue') # specify color by name
plt.plot(x,np.sin(x-1),color = 'g') # (rgbcmyk)
plt.plot(x,np.sin(x-2),color = '0.75')# Grayscale
plt.plot(x,np.sin(x-3),color = '#FFDD44') #Hex code (RRGGBB from 00 to FF)
plt.plot(x,np.sin(x-4),color = (1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x,np.sin(x-5),color = 'chartreuse'); # all HTML colors suported
```

```python
# linestyles
linestyles = ['solid','dashed','dashdot','dotted']
translation = [0,1,2,3]
for ls,t in zip(linestyles,translation):
    plt.plot(x,x+t,linestyle = ls)
# alternative way
linestyles = ['-','--','-.',':']
translation = [4,5,6,7]
for ls,t in zip(linestyles,translation):
    plt.plot(x,x+t,linestyle = ls)
```

```python
# linestyle and color in the same command!
plt.plot(x,x+0,'-g') # solid green
plt.plot(x,x+1,'--c')# dashed cyan
plt.plot(x,x+2,'-.k') # dashdot black
plt.plot(x,x+3,':r') #dotted red
```

## Continuous Errors

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
kernel =  RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)
gp.fit(xdata[:, np.newaxis], ydata)
xfit = np.linspace(0, 10, 1000)
y_pred, sigma = gp.predict(xfit.reshape(-1,1), return_std=True)
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, y_pred, '-', color='gray')
plt.fill_between(xfit, y_pred - sigma, y_pred + sigma,
color='gray', alpha=0.2)
plt.xlim(0, 10);
```

## Visualization of Three-Dimensional Function

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
plt.rcParams["axes.grid"] = False
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='YlGnBu');
```

```python
# To check another colormaps type
#plt.cm.<TAB>!
```

```python
# Fill the counour with plt.contourf
plt.contourf(X, Y, Z, 20, cmap='YlGnBu');
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='upper',
cmap='RdGy')
plt.colorbar()
plt.axis('image');
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=False, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
cmap='RdGy', alpha=0.5)
plt.colorbar();
```

## Two-Dimensional Histogram and Binnings

```python
# Setting a multivariate normal
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
kwargs = dict(bins = 30,cmap = 'Greens')#,density = True)
plt.hist2d(x, y, **kwargs)
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
# using numpy 
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
plt.hexbin(x, y, gridsize=30, cmap='Greens')
cb = plt.colorbar(label='count in bin')
```

## Kernel Density Estimation (KDE)

```python
from scipy.stats import gaussian_kde
```

```python
# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)
# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
```

```python
# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
origin='lower', aspect='auto',
extent=[-3.5, 3.5, -6, 6],
cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```
